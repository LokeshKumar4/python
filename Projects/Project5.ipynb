# Problem Statement 20: (Run Query on the database and use in python)
Write a query that finds the number of households without a toilet on the premise for each district of the state that has the least hospital bed-to-population ratio. 
Display the information using python.



```python
query='''
create view service_less as SELECT h.`State/UT`,h.District,
(h.Households_Rural+h.Households_Urban - (h.Households_Rural_Toilet_Premise + h.Households_Urban_Toilet_Premise)) as Household_without_toilet_inpremises  ,
(h.Population*0.003 - (a.HospitalBeds*1000 / h.Population)) as gap_of_beds FROM housing h 
join all_hospitals a 
on h.`State/UT`=a.`State/UT`
where (h.Population*0.003 - (a.HospitalBeds*1000 / h.Population))>0 and 
((h.Households_Rural+h.Households_Urban) - (h.Households_Rural_Toilet_Premise + h.Households_Urban_Toilet_Premise))>0
order by gap_of_beds ;

'''
```

```python
c.execute(query)
```

```python
c.execute("select count(distinct(`State/UT`)) from service_less;")
for val in c:
    print(val)
```

# Problem Statement 21: 
Create stored functions for the following:

Stored Function name -->	Input parameter	-->        Return parameter

get_population_district	-->District Name-->	Total population of the district

get_population	-->State/UT Name-->	Total population of the district

senior_citizen_population -->	State/UT Name	-->Total Senior citizens in the state

get_hospital_beds-->	State/UT Name-->	Total Number of hospital beds in the State/UT

get_govt_hospital_beds-->	State/UT Name-->	Total Number of government hospital beds in the State/UT

beds_per_lakh-->	State/UT Name-->	Total Number of hospital beds per 1 lakh people

govt_beds_per_lakh-->	State/UT Name-->	Total Number of government hospital beds per 1 lakh people


```python
c.execute('''
CREATE PROCEDURE if not exists get_population_district (in dis varchar(48))
BEGIN
    SELECT District, Population FROM census c where c.District = dis ;
END ; 
''')
```

```python
c.execute('''call get_population_district('Agra');''')
for val in c:
    print(val)
```

```python
c.execute('''
CREATE PROCEDURE if not exists get_population (in state varchar(48))
BEGIN
    SELECT `State/UT`,District,Population FROM census c where c.`State/UT` = state ;
END ; 
''')
```

```python
c.execute('''call get_population('Bihar');''')
for val in c:
    print(val)
```

```python
c.execute('''
CREATE PROCEDURE if not exists senior_citizen_population (in state varchar(40))
BEGIN
    SELECT `State/UT`, sum(Senior_Citizen) as Total_Senior_citizen FROM census c where c.`State/UT` = state
    group by `State/UT`;
END ; 
''')
```

```python
c.execute('''call senior_citizen_population('Bihar');''')
for val in c:
    print(val)
```

```python
# get_hospital_beds-->	State/UT Name-->	Total Number of hospital beds in the State/UT
c.execute('''
CREATE PROCEDURE if not exists get_hospital_beds (in state varchar(40))
BEGIN
    SELECT `State/UT`, HospitalBeds FROM all_hospitals a where a.`State/UT` = state ;
END ; 
''')
```

```python
c.execute('''call get_hospital_beds('Bihar');''')
for val in c:
    print(val)
```

```python
# get_govt_hospital_beds-->	State/UT Name-->	Total Number of government hospital beds in the State/UT
c.execute('''
CREATE PROCEDURE if not exists get_govt_hospital_beds (in state varchar(40))
BEGIN
    SELECT `State/UT`, Rural_Government_Beds+ Urban_Government_Beds as Total_beds FROM government_hospital g where g.`State/UT` = state ;
END ; 
''')
```

```python
c.execute('''call get_govt_hospital_beds('Bihar');''')
for val in c:
    print(val)
```

```python
# beds_per_lakh-->	State/UT Name-->	Total Number of hospital beds per 1 lakh people
c.execute('''
CREATE PROCEDURE beds_per_lakh (in state varchar(40))
                    BEGIN
                        with ct1 as  (SELECT c.`State/UT`, sum(c.Population) as Population from census c group by 
                            c.`State/UT` )
                        select ct1.`State/UT`, (a.HospitalBeds/ct1.Population)*100000 as Beds_per_1Lakh from 
                        ct1 join all_hospitals a on ct1.`State/UT` = a.`State/UT` where ct1.`State/UT` = state ;  
                    END 
''')
```

```python
c.execute('''call beds_per_lakh('Bihar');''')
for val in c:
    print(val)
```

```python
c.execute('''
CREATE PROCEDURE govt_beds_per_lakh (in state varchar(40))
                    BEGIN
                        with ct1 as  (SELECT c.`State/UT`, sum(c.Population) as Population from census c group by 
                        c.`State/UT` )
                        select ct1.`State/UT`, ((g.Urban_Government_Beds + g.Rural_Government_Beds)/ct1.Population)*100000 
                        as Beds_per_1Lakh from 
                        ct1 join government_hospital g on ct1.`State/UT` = g.`State/UT` where ct1.`State/UT` = state ;  
                    END ;''')
```

```python
c.execute('''call govt_beds_per_lakh('Bihar');''')
for val in c:
    print(val)
```
