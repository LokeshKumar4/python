```python
def func_rename(a):
    if "&" in a:
        v=a.split("&")
        st=v[0].strip().title() +" And " +v[1].strip().title()
        return st
    if "*" in a:
        v=a.replace("*","")
        return v.strip().title()
    return a.title()
```

```python
df_copy['State/UT']=df_copy['State/UT'].apply(func_rename)
```

# Problem Statement 15: (Government healthcare facility disparity)
Since the resources are limited, it is required to identify the region which lacks the healthcare facility the most for creating new government hospitals.
An idea was suggested* that first the three States/UTs which have the least amount of beds (in all hospitals government and private) for their population is identified. Among those three the state which has the least number of government hospitals can be recommended for setting up a new government hospital.


```python
df_housing=pd.read_csv("clean_data/housing.csv",usecols=["State/UT","Population"])
df_housing["State/UT"]=df_housing['State/UT'].apply(func_rename)
df_housing
```

```python
df_housing=df_housing.groupby("State/UT").sum()
```

```python
# df_housing=pd.DataFrame(df_housing)
df_housing
```

```python
df_combine=df_copy.merge(df_housing,left_on="State/UT",right_on="State/UT",how='left')
```

```python
df_combine.info()
```

```python
df_combine['Rural_Government_Beds']=df_combine['Rural_Government_Beds'].astype('float')
```

```python
df_combine['Urban_Government_Beds']=df_combine['Urban_Government_Beds'].astype('float')
```

```python
df_combine
```

```python
df_total=pd.read_csv("clean_data/all_hospitals.csv",usecols=['State/UT','Hospitals','HospitalBeds'] )
df_total
df_combine_hospital=df_combine.merge(df_total,on="State/UT")
df_combine_hospital
```

```python
df_combine_hospital['HospitalBeds']=df_combine_hospital['HospitalBeds'].astype('float')
df_combine_hospital.info()
```

```python
df_combine_hospital["beds_per_1000_person"]=df_combine_hospital["HospitalBeds"] *1000 / df_combine_hospital["Population"]
df_combine_hospital
```

```python
df_combine_hospital.loc[df_combine_hospital["beds_per_1000_person"]<3]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
           
plt.bar(x=df_combine_hospital['State/UT'],height=df_combine_hospital['beds_per_1000_person'],color="skyblue")
plt.ylabel("Beds in Hospital")
plt.xlabel("States")
plt.title("Beds in each state per 1000 person")
plt.xticks(rotation = 90)

plt.ylim(0,5)
```

# Problem Statement 16: (Gap in number of beds)
Visually represent the difference between the expected number of hospital beds and the available number of hospital beds in each State/UT, as well as at the national level (if any).


```python
df_combine_hospital['Expected_Beds']=df_combine_hospital['Population']*0.003

df_combine_hospital['Beds_Required']=df_combine_hospital['Expected_Beds']-df_combine_hospital['HospitalBeds']
df_combine_hospital
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_combine_hospital['State/UT'],height=3,label="Expected Beds",color="skyblue")
plt.bar(x=df_combine_hospital['State/UT'],height=df_combine_hospital['beds_per_1000_person'],color="green",label="Actual Beds Available")
plt.ylabel("Beds in Hospital")
plt.xlabel("States")
plt.title("Beds in each state per 1000 person")
plt.xticks(rotation = 90)
plt.legend()
plt.ylim(0,)
```

# Problem Statement 17: (Hospitals required to meet the standards)
In a hypothetical situation if the government wishes to fill the gap between the available facility and the WHO standards all by itself how many government hospitals may be required in each state or union territory as well as at the national level?
Find the average number of beds in a government hospital. Divide it by the gap in the number of beds to reach the standards in the region, and round the number to the nearest integer to find the value.
Represent the findings visually.


```python
df_hospital_data=df_combine_hospital.copy()
df_hospital_data
```

```python
df_hospital_data['Rural_Government_Hospitals']=df_hospital_data['Rural_Government_Hospitals'].astype('float')
df_hospital_data['Urban_Government_Hospitals']=df_hospital_data['Urban_Government_Hospitals'].astype('float')

```

```python
df_hospital_data['average_gov_hospitals_bed']=((df_hospital_data['Rural_Government_Beds']+ df_hospital_data['Urban_Government_Beds'])/(df_hospital_data['Rural_Government_Hospitals']+df_hospital_data['Urban_Government_Hospitals']))
df_hospital_data
```

```python
df_hospital_data['Hospitals_Required']=round(df_hospital_data['Beds_Required']/ (df_hospital_data['average_gov_hospitals_bed']))

df_hospital_data
```

```python
plt.figure(figsize=(10,6))
           
plt.bar(x=df_hospital_data['State/UT'],height=df_hospital_data['Hospitals_Required'],color=["skyblue"])
plt.ylabel("Hospitals Required")
plt.xlabel("State/UT")
plt.title("Hospital expension required as per WHO")
plt.xticks(rotation = 90)

plt.ylim(0,)
```
