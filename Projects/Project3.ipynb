# Problem Statement 11: 

(Create a function to alter the data to create uniformity)
The State/UT names have some different formats compared to other census and housing data. Manually identify those differences and fix them so that the State/UT names match exactly with the census and housing data. 
A function should be created to perform this operation. Since the same operation is required for another dataset as well. 
After the process save the data in a CSV file named all_hospitals.csv in the Clean_Data folder



```python
df
```

```python
def func_alter(a):
    if "&" in a:
        a1=a.split("&")
        st=str(a1[0].strip().title()+" And "+ a1[1].strip().title())
        return st
    elif a=="Odisha":
        return "Orissa".title()
    elif a=="Puducherry":
        return "Pondicherry".title()
    elif a=="Delhi":
        return "Nct Of Delhi".title()
    else:
        return a.strip().title()
```

```python
df["State/UT"]=df["State/UT"].apply(func_alter)
```

```python
df
```

```python
df2=pd.read_csv("clean_data/housing.csv",usecols=["State/UT","Population"])
df2['State/UT']=df2['State/UT'].apply(func_alter)
```

```python
df2_group=df2.groupby('State/UT').sum()
df2_group.head(2)
df3=pd.DataFrame( df2_group)
df3
```
